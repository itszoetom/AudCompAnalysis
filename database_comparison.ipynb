```python
import os

import pandas as pd
from scipy import stats
from jaratoolbox import celldatabase
from jaratoolbox import histologyanalysis as ha
from jaratoolbox import spikesorting
from jaratoolbox import settings
import database_generation_funcs as funcs
import params
import sys
```

```python
subjects = params.subject_list
df = pd.DataFrame({})
fullPath = '/Volumes/NardociData/old_databases'

for subject in subjects:
    dbFilename = os.path.join(fullPath, f'celldb_{subject}.h5')
    cellDb = celldatabase.load_hdf(dbFilename)
    df = pd.concat([df, cellDb])
    
simpleSiteNames = df['recordingSiteName'].str.split(',').apply(lambda x: x[0])
df['recordingSiteName'] = simpleSiteNames
```
